```python
from choroidalyze import Choroidalyzer

# This initialises choroidalyzer. 
# It will try to automatically download the model weights from github the first time you run it. 
choroidalyzer = Choroidalyzer()

# basic useage: get the metrics
metrics = choroidalyzer.analyze('example_data/image1.png')
print(metrics)
```

```python
metrics = choroidalyzer.analyze('example_data/image2.png')
print(metrics)
```

```python
metrics = choroidalyzer.analyze('example_data/image3.png')
print(metrics)
```

```python
# advanced usage: specify the pixel scaling values. These can be extracted from the image meta data if available
# choroidalyzer uses a default value of (11.49, 3.87) if no value is specified
metrics = choroidalyzer.analyze('example_data/image1.png', scale=(11.49, 3.87))
print(metrics)

metrics = choroidalyzer.analyze('example_data/image1.png', scale=(11.02, 3.87))
print(metrics)

metrics = choroidalyzer.analyze('example_data/image1.png', scale=(12.75, 3.87))
print(metrics)
```

```python
# choroidalyzer also has a basic plotting function to inspect segmentation outputs
choroidalyzer.predict_and_plot('example_data/image1.png')
```

```python
choroidalyzer.predict_and_plot('example_data/image2.png')
```

```python
choroidalyzer.predict_and_plot('example_data/image3.png')
```

```python
# you can also access just the segmentations
raw_segmentations = choroidalyzer.predict('example_data/image1.png')
print('The segmentation has 3 channels: first channel is region, second vessels, third fovea')
print(choroidalyzer.outputs)
print(raw_segmentations.shape)
print('Note that the segmentations by default are not binarized, so you can access the raw pixel-wise probabilities')
```
